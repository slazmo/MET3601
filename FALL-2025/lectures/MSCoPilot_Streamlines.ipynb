```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters for Problem 1
e = 2
g = 1

# Grid setup
#DRL I had to edit this from the AI interpretation since it grid only went from (0, 16)!
x, y = np.meshgrid(np.arange(0, 20), np.arange(0, 20))

# Velocity components
u = e * x - g * y
v = g * x - e * y
# Streamline starting points
sx = [2.5, 5, 7.5, 10, 12.5, 2.5, 3.5, 4.5, 5.5, 6.5, 7.5, 8.5, 9.5, 10.5, 11.5, 12.5]
sy = [5.5, 1, 1, 1, 1, 16, 16, 16, 16, 16, 16, 16, 16, 16, 16, 16]
```

```python
# Plotting
plt.figure(figsize=(8, 8))
plt.streamplot(x, y, u, v, start_points=np.column_stack((sx, sy)), color='b')
plt.quiver(x, y, u, v, color='r')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('PROB #1: 2D FLOW W/ STREAMLINES')
plt.axis('equal')
plt.grid(True)
plt.show()
```
